# Face Recognition from an image

```python
import cv2,matplotlib.pyplot as plt
img=plt.imread(r"..\Aditya\.anaconda\navigator\images\friends.jpg")

face_cascade=cv2.CascadeClassifier(r'..\Aditya\.anaconda\navigator\DATA\haarcascades\haarcascade_frontalface_default.xml')
def detect_face(img):
    face_img=img.copy()
    face_rects=face_cascade.detectMultiScale(face_img)
    
    for x,y,w,h in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,0,0),3)
    
    return face_img

result=detect_face(img)
plt.imshow(result)
plt.show()
```

# Face Recognition from WebCam

```python
cap=cv2.VideoCapture(0)

face_cascade=cv2.CascadeClassifier(r'..\Aditya\.anaconda\navigator\DATA\haarcascades\haarcascade_frontalface_default.xml')
def detect_face(img):
    face_img=img.copy()
    face_rects=face_cascade.detectMultiScale(face_img)
    
    for x,y,w,h in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,0,0),3)
    
    return face_img

if cap.isOpened:
    ret,frame=cap.read()
else:
    ret=False
    
while ret:
    ret,frame=cap.read()
    
    cv2.imshow("Camera",detect_face(frame))
    if cv2.waitKey(1)==27:
        break
        
cv2.destroyAllWindows()
cap.release()
```
